```python
df_openinsider = pd.read_csv('output_openinsider.csv')
```

```python
df_recommendations = pd.read_excel('analyst_recommendations.xlsx',header=1)
```

```python
rating_mapping = {
    'Buy': 'Buy',
    'Neutral': 'Hold',
    'Overweight': 'Buy',
    'Outperform': 'Buy',
    'Hold': 'Hold',
    'Equal-Weight': 'Hold',
    'Market Perform': 'Hold',
    'Underweight': 'Sell',
    'Underperform': 'Sell',
    'Sell': 'Sell',
    'Sector Perform': 'Hold',
    'Strong Buy': 'Buy',
    'In-Line': 'Hold',
    'Positive': 'Buy',
    'Market Outperform': 'Buy',
    'Sector Weight': 'Hold',
    'Perform': 'Hold',
    'Equal-weight': 'Hold',
    'Reduce': 'Hold',
    'Sector Outperform': 'Buy',
    'Peer Perform': 'Hold',
    'Accumulate': 'Buy',
    'Negative': 'Sell',
    'Long-Term Buy': 'Buy',
    'Top Pick': 'Buy',
    'Fair Value': 'Hold',
    'Mixed': 'Other',
    'Conviction Buy': 'Buy',
    'Market Weight': 'Hold',
    'Sector Underperform': 'Sell',
    'Outperformer': 'Buy',
    'Market Underperform': 'Sell',
    'Strong Sell': 'Sell',
    'Add': 'Buy',
    'Long-term Buy': 'Buy',
    'Cautious': 'Hold',
    'Average': 'Other',
    'Trim': 'Other',
    'Above Average': 'Other',
    'Below Average': 'Other',
    'Underperformer': 'Sell',
    'In-line': 'Hold',
    'Speculative Buy': 'Buy',
    'Peer perform': 'Hold',
    'Hold Neutral': 'Hold',
    'Gradually Accumulate': 'Hold',
}
```

```python
df_recommendations['Rating_Category'] = df_recommendations['Rating'].map(rating_mapping).fillna('Other')
```

```python
df_recommendations['Stock_vs_SP500_1week'] = df_recommendations['% change 1 week stock price'] - df_recommendations['% change 1 week SPY price']
df_recommendations['Stock_vs_SP500_4week'] = df_recommendations['% change 4 week stock price'] - df_recommendations['% change 4 week SPY price']
df_recommendations['Stock_vs_SP500_12week'] = df_recommendations['% change 12 week stock price'] - df_recommendations['% change 12 week SPY price']
```

```python
df_recommendations['Date'] = pd.to_datetime(df_recommendations['Date'])  # Convert 'date' to datetime format

# Create new columns for year, month, and day
df_recommendations['Year'] = df_recommendations['Date'].dt.year
df_recommendations['Month'] = df_recommendations['Date'].dt.month
df_recommendations['Day'] = df_recommendations['Date'].dt.day
```

```python
df_recommendations = df_recommendations[['Company Name','Ticker','Date','Stock Price on day of recommendation','Stock Price after 4 weeks','Rating_Category','Year','Month','Day']]
```

```python
df_recommendations.to_csv('output_recommendations.csv')
```
