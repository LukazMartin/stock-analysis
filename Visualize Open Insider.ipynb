```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime, timezone
```

```python
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")
```

```python
# Function to fetch historical stock prices
def get_historical_stock_prices(ticker, start_date, end_date):
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    return stock_data['Adj Close']
```

```python
df = pd.read_csv('open_insider.csv')

# Filter rows for AAPL ticker
aapl_df = df[df['company_name'] == 'AAPL']

# Convert 'trade_date' to string and then back to datetime to avoid timezone issues
aapl_df['trade_date'] = pd.to_datetime(aapl_df['trade_date'].astype(str))

# Fetch historical stock prices for AAPL
start_date = aapl_df['trade_date'].min() - pd.DateOffset(years=10)  # Assuming 10 years of historical data
end_date = datetime.today().strftime('%Y-%m-%d')
aapl_prices = get_historical_stock_prices('AAPL', start_date, end_date)
```

```python
df['transaction_type_category'] = df['transaction_type'].apply(lambda x: 1 if x == 'P - Purchase' else 0)
```

```python
df['transaction_type_category'].value_counts()
```

```python
def get_market_cap(ticker):
    try:
        stock_info = yf.Ticker(ticker)
        return stock_info.info['marketCap']
    except Exception as e:
        print(f"Error fetching market cap for {ticker}: {e}")
        return None
```

```python
# Create a DataFrame with unique tickers
unique_tickers = pd.DataFrame(df['company_name'].unique(), columns=['company_name'])
```

```python
# Fetch market cap for each unique ticker
unique_tickers['market_cap'] = unique_tickers['company_name'].apply(get_market_cap)
```

```python
# Merge market cap information back into the original DataFrame
df = pd.merge(df, unique_tickers, on='company_name', how='left')
```

```python
# Create a new column 'percentage_of_market_cap'
df['percentage_of_market_cap'] = 0.0  # Initialize with 0
```

```python
# Convert 'Qty' column to numeric (removing commas)
df['Qty'] = pd.to_numeric(df['Qty'].str.replace(',', ''), errors='coerce')

# Convert 'last_price' column to numeric
df['last_price'] = pd.to_numeric(df['last_price'].str.replace('$', ''), errors='coerce')

# Ensure 'market_cap' is numeric
df['market_cap'] = pd.to_numeric(df['market_cap'], errors='coerce')
```

```python
# Update the values based on the transaction type
df.loc[df['transaction_type_category'] == 1, 'percentage_of_market_cap'] = (
    df['Qty'] * df['last_price'] / df['market_cap']
)
```

```python
df.loc[df['transaction_type_category'] == 0, 'percentage_of_market_cap'] = (
    -df['Qty'] * df['last_price'] / df['market_cap']
)
```

```python
# Drop rows where 'market_cap' is NaN
df.dropna(subset=['market_cap'], inplace=True)

# Reset the index if needed
df.reset_index(drop=True, inplace=True)
```
