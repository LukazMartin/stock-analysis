```python
# Function to fetch historical stock prices
def get_historical_stock_prices(ticker, start_date, end_date):
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    return stock_data['Adj Close']
```

```python

# Filter rows for AAPL ticker
aapl_df = df[df['company_name'] == 'AAPL']

# Convert 'trade_date' to string and then back to datetime to avoid timezone issues
aapl_df['trade_date'] = pd.to_datetime(aapl_df['trade_date'].astype(str))

# Fetch historical stock prices for AAPL
start_date = aapl_df['trade_date'].min() - pd.DateOffset(years=10)  # Assuming 10 years of historical data
end_date = datetime.today().strftime('%Y-%m-%d')
aapl_prices = get_historical_stock_prices('AAPL', start_date, end_date)

# Plot the historical stock prices
plt.figure(figsize=(10, 6))
plt.plot(aapl_prices.index, aapl_prices, label='AAPL Historical Stock Price')

# Round the timestamps in the DataFrame and historical prices to match
aapl_df['rounded_trade_date'] = aapl_df['trade_date'].dt.floor('D')
aapl_prices.index = aapl_prices.index.floor('D')

# Add red points for trade dates
for index, row in aapl_df.iterrows():
    try:
        plt.scatter(row['rounded_trade_date'], aapl_prices.loc[row['rounded_trade_date']], color='red', label='Trade Date')
    except:
        pass

# Set plot labels and title
plt.xlabel('Date')
plt.ylabel('Stock Price ($)')
plt.title('AAPL Historical Stock Price with Trade Dates')


plt.xticks([], [])
plt.yticks([], [])

# Show the plot
plt.show()
```
